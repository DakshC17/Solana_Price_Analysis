### DATA FIXING

## Checking for Date Time columns and also check is numeric columns contains numeric values...

```python
df.dtypes
```

```python
df['Open time'] = pd.to_datetime(df['Open time'])
df['Close time'] = pd.to_datetime(df['Close time'])
```

```python
df = df.sort_values('Open time')
df = df.set_index('Open time')
```

```python
df.info()
df.head()
```

```python
df.isnull().sum()
```

```python
df.index.duplicated().sum()
```

```python

df['price_change'] = df['Close'] - df['Open']

df['return_pct'] = (df['Close'] - df['Open']) / df['Open'] * 100

df['range'] = df['High'] - df['Low']

df['volatility_pct'] = (df['High'] - df['Low']) / df['Open'] * 100
```

```python
df.head()
```

```python

df['day_name'] = df.index.day_name()

df['day_number'] = df.index.dayofweek

df['month'] = df.index.month

df['year'] = df.index.year
```

```python
df.head()
```

```python
weekday_returns = df.groupby('day_name')['return_pct'].mean().sort_values()

weekday_returns
```

```python
import matplotlib.pyplot as plt

weekday_returns.plot(kind='bar', figsize=(8,4))
plt.title("Average Return % by Day")
plt.ylabel("Return %")
plt.show()
```

```python
weekday_volatility = df.groupby('day_name')['volatility_pct'].mean().sort_values()

weekday_volatility

weekday_volatility.plot(kind='bar', figsize=(8,4), color='orange')
plt.title("Average Volatility % by Day")
plt.ylabel("Volatility %")
plt.show()
```

```python
monthly_returns = df.groupby('month')['return_pct'].mean()

# monthly_returns.plot(kind = 'bar', figsize=(8,4), color = "green")
monthly_returns.plot(figsize=(8,4))
plt.title("Monthly Average Returns")
plt.ylabel("Return %")
plt.show()
```

```python
positive_days = (df['return_pct'] > 0).sum()
negative_days = (df['return_pct'] <= 0).sum()

positive_days, negative_days
```

```python
big_moves = df[abs(df['return_pct']) > 10]

big_moves[['Close','return_pct']].head(10)
```

```python
df.head()
```

```python
df["ma_7"] = df["Close"].rolling(window=7).mean()
```

```python
df["ma_30"] = df["Close"].rolling(window=30).mean()
```

```python
df["ma_90"] = df["Close"].rolling(window=90).mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df['Close'], label='Price', alpha=0.5)
plt.plot(df['ma_7'], label='7D Trend')
plt.plot(df['ma_30'], label='30D Trend')
plt.plot(df['ma_90'], label='90D Trend')

#Plotting has been done for the trend according to the days

plt.legend()
plt.title("Solana Trends")
plt.show()
```

```python
df["market_state"] = ["Bull" if x > y else "Bear" for x , y in zip(df["ma_7"],df["ma_30"])]
df[['Close','ma_7','ma_30','market_state']].tail(10)
```
